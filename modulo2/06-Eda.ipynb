# EDA

### **Exploratory Data Analysis (EDA)** es el proceso de investigar, resumir y visualizar un dataset antes de aplicar modelos o tomar decisiones.

### Objetivos principales del EDA:

- Comprender la estructura y características del dataset.
- Detectar valores faltantes, outliers y errores.
- Analizar la distribución y relaciones entre variables.
- Generar visualizaciones que ayuden a interpretar los datos.

## 🚢 Introducción al Dataset del Titanic

El dataset del Titanic es uno de los más utilizados para enseñanza de Análisis de Datos y Machine Learning. Contiene información sobre los pasajeros del Titanic, su perfil y si sobrevivieron o no al hundimiento.

| Columna    | Tipo       | Descripción                                             |
| ---------- | ---------- | ------------------------------------------------------- |
| `Survived` | Numérica   | 0 = No sobrevivió, 1 = Sobrevivió                       |
| `Pclass`   | Numérica   | Clase del ticket: 1 = Primera, 2 = Segunda, 3 = Tercera |
| `Sex`      | Categórica | Sexo del pasajero: male/female                          |
| `Age`      | Numérica   | Edad del pasajero (años)                                |
| `Fare`     | Numérica   | Tarifa pagada por el ticket                             |

# 2 Valores faltantes y duplicados

```python
# Valores faltantes
print(df.isna().sum())
```

```python
# Rellenar valores faltantes con la media de Age
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
# Duplicados
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

# 3. Análisis Univariado

## 3.1 Distribución de variables numéricas

```python
# Histograma de Age
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
plt.hist(df['Age'], bins=40, color='skyblue', edgecolor='black')
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))  # tamaño compacto
sns.histplot(
    df['Age'],
    bins=20,
    kde=True,               # curva de densidad
    color="#69b3a2",        # color moderno
    edgecolor="black",
    alpha=0.7               # transparencia
)

# Mejorar títulos y etiquetas
plt.title('Distribución de Edad', fontsize=12, fontweight='bold')
plt.xlabel('Edad', fontsize=10)
plt.ylabel('Frecuencia', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.grid(alpha=0.3)        # líneas de grilla suaves
plt.tight_layout()
plt.show()
```

```python
# Histograma de Fare
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
plt.hist(df['Fare'], bins=20, color='orange', edgecolor='black')
plt.title('Distribución de Fare')
plt.xlabel('Fare')
plt.ylabel('Frecuencia')
plt.show()
```

## 3.2 Distribución de variables categóricas

```python
# Survived
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
df['Survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title('Supervivencia')
plt.xlabel('0 = No, 1 = Sí')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Sex
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
df['Sex'].value_counts().plot(kind='bar', color=['blue','pink'])
plt.title('Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Pclass
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
df['Pclass'].value_counts().sort_index().plot(kind='bar', color='gray')
plt.title('Clase de pasajero')
plt.xlabel('Pclass')
plt.ylabel('Cantidad')
plt.show()
```

# 4. Análisis Bivariado

```python
# Relación Age vs Survived
plt.figure(figsize=(12,3))  # ancho=4, alto=3 pulgadas
plt.scatter(df['Age'], df['Survived'], alpha=0.5)
plt.xlabel('Edad')
plt.ylabel('Survived')
plt.title('Edad vs Supervivencia')
plt.show()
```

```python
# Boxplot Age por Survived

fig, ax = plt.subplots(figsize=(4,5))  # ancho=4, alto=3 pulgadas

# Boxplot usando el eje ax
df.boxplot(column='Age', by='Survived', ax=ax)

# Títulos
ax.set_title('Edad por Supervivencia', fontsize=10)
ax.set_xlabel('Survived', fontsize=8)
ax.set_ylabel('Age', fontsize=8)
plt.suptitle('')  # Elimina el título automático
plt.tight_layout()
plt.show()
```

```python
# Fare por Survived

fig, ax = plt.subplots(figsize=(4,3))  # ancho=4, alto=3 pulgadas

# Boxplot Fare vs Survived
df.boxplot(column='Fare', by='Survived', ax=ax)

# Títulos
ax.set_title('Fare por Supervivencia', fontsize=10)
ax.set_xlabel('Survived', fontsize=8)
ax.set_ylabel('Fare', fontsize=8)
plt.suptitle('')  # Elimina el título automático
plt.tight_layout()
plt.show()
```

```python
# Comparación de categorías
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
df.groupby('Sex')['Survived'].mean().plot(kind='bar', color=['blue','pink'])
plt.title('Tasa de supervivencia por sexo')
plt.ylabel('Proporción')
plt.show()

plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
df.groupby('Pclass')['Survived'].mean().plot(kind='bar', color='gray')
plt.title('Tasa de supervivencia por clase')
plt.ylabel('Proporción')
plt.show()
```

# 5. Correlaciones

```python
# Matriz de correlación
corr = df.select_dtypes(include=np.number).corr()
print(corr)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

```python
# Outliers en Fare
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)
iqr = q3 - q1

outliers_fare = df[(df['Fare'] < q1 - 1.5*iqr) | (df['Fare'] > q3 + 1.5*iqr)]
print(outliers_fare)
```

```python
# Pivot tables y resumen por categoría
# Tasa de supervivencia por Pclass y Sex
pivot = df.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean')
print(pivot)

# Visualización
plt.figure(figsize=(4,3))  # ancho=4, alto=3 pulgadas
pivot.plot(kind='bar')
plt.title('Tasa de supervivencia por clase y sexo')
plt.ylabel('Proporción')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Colores por sexo
colors = df['Sex'].map({'male':'blue','female':'pink'})

# Tamaño según supervivencia
sizes = df['Survived']*50 + 20  # sobrevivientes más grandes

plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'], c=colors, s=sizes, alpha=0.6, edgecolor='k')
plt.xlabel('Edad')
plt.ylabel('Fare')
plt.title('Edad vs Fare (Color=Sexo, Tamaño=Survived)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar variables numéricas + codificación de Sex
df['Sex_num'] = df['Sex'].map({'male':0, 'female':1})
pair = sns.pairplot(
    df[['Survived','Pclass','Age','Fare','Sex_num']], 
    hue='Survived', 
    diag_kind='kde', 
    palette='coolwarm',
    height=2 
)

# Título de la figura
pair.fig.suptitle('Relaciones entre variables numéricas', y=1.02)  # y=1.02 para separarlo
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necesario para 3D

# Crear figura 3D
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

# Ejes
x = df['Age']
y = df['Fare']
z = df['Pclass']

# Color por supervivencia
colors = df['Survived'].map({0:'red', 1:'green'})

# Scatter 3D
scatter = ax.scatter(x, y, z, c=colors, s=50, alpha=0.7, edgecolor='k')

# Etiquetas
ax.set_xlabel('Edad')
ax.set_ylabel('Fare')
ax.set_zlabel('Pclass')
ax.set_title('Titanic 3D: Age vs Fare vs Pclass (Color=Survived)')

# Leyenda manual
from matplotlib.lines import Line2D
legend_elements = [Line2D([0],[0], marker='o', color='w', markerfacecolor='red', label='No sobrevivió', markersize=8),
                   Line2D([0],[0], marker='o', color='w', markerfacecolor='green', label='Sobrevivió', markersize=8)]
ax.legend(handles=legend_elements, loc='upper right')

plt.show()
```

# Conclusiones del EDA

- La edad promedio está alrededor de 30 años, con algunos valores muy bajos y mayores de 70.

- Los valores faltantes de Age fueron completados con la media.

- La supervivencia es mayor en mujeres que hombres y en pasajeros de primera clase.

- Fare tiene algunos outliers muy altos (personas con tickets muy caros).

- La matriz de correlación indica que Fare y Pclass tienen una relación moderada con Survived.
