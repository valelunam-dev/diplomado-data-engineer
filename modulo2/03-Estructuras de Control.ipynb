# Condicionales (`if`, `elif`, `else`)

En Python podemos **tomar decisiones** usando estructuras condicionales.  
La estructura básica es:

```python
if condición:
    # código si la condición es verdadera
elif otra_condición:
    # código si la condición anterior no se cumple, pero esta sí
else:
    # código si ninguna condición anterior se cumple


### Ejemplo 1: `if`

````python
edad = 452

if edad >= 18:
	print("Eres mayor de edad")
print("Fin del programa")
````

### Ejemplo 2: `if` + `else`

````python
edad = 15

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
````

### Ejemplo 3: `if` + `elif` + `else`

````python
nota = 2

if nota >= 6:
    print("Excelente")
elif nota >= 4:
    print("Aprobado")
else:
    print("Reprobado")
````

### Ejemplo 4: Varios `elif`

````python
dia = 18

if dia == 1:
    print("Lunes")
elif dia == 2:
    print("Martes")
elif dia == 3:
    print("Miércoles")
else:
    print("Otro día")
````

### Ejemplo 5: Condicional con operadores lógicos

Operadores lógicos

Operador | Nombre | Ejemplo | Resultado |
:--------: | :-------:| :--------: | :------------------------:
**&** | y (and)| False & True | Devuelve **False**. Si ambos son True, devuelve True
**\|** | o (or)| False \| True | Devuelve **True**. Si al menos uno de True devuelve True
**!=** | distinto de (not)| True != True| Devuelve **False**. Devuelve lo opuesto al resultado de la evaluación

````python
nota= 4
asistencia = 77

if (nota >= 4 and asistencia >= 75):
    print("Aprobado")
else:
    print("Reprobado")
````

### Ejemplo 6: Condicional con múltiples condiciones

````python
temperatura = 15
humedad = 80

if temperatura > 30:
    if humedad > 70:
        print("Día muy caluroso y húmedo")
    else:
        print("Día caluroso")
elif 20 <= temperatura <= 30:
    print("Día agradable")
else:
    print("Día frío")
````

### If y Elif
Hay una diferencia fundamental, si usas if encadenados todas las expresiones son evaluadas siempre, por el contrario si usas elif, en el momento que una condición anterior (el if u otro elif se cumpla), ese elif no se evalúa siquiera.Ejemplo

````python
#Con if
a = 15
if a % 5 == 0:
    print(f"{a} es divisible entre 5")
if a % 3 == 0:
    print(f"{a} es divisible entre 3")
````

````python
#Con elif
a = 12

if a % 5 == 0:
    print(f"{a} es divisible entre 5")
elif a % 3 == 0:
    print(f"{a} es divisible entre 3")
````

# 2 Bucles (for, while)

## 2.1 Bucles `for` en Python

El bucle `for` en Python nos permite **recorrer elementos de una secuencia** (listas, cadenas, rangos, diccionarios, etc.) y ejecutar un bloque de código para cada elemento.

---

### Sintaxis

```python
for variable in secuencia:
    # código a ejecutar en cada iteración


### Ejemplo 1: Recorrer una lista

````python
ropas = ["pantalon", "polera", "zapato"]

for ropa in ropas:
    print(ropa)
````

### Ejemplo 2: Usando range()

````python
for i in range(3):
    print("Iteración:", i)
````

### Ejemplo 3: range(inicio, fin, paso)

````python
for i in range(2, 11):
    print(i)
````

### Ejemplo 4: Recorrer una cadena de texto

````python
palabra = "Python"

for letra in palabra:
    print(letra)
````

### Ejemplo 5: for con enumerate()

A veces necesitamos índice + valor al recorrer una lista. Para eso usamos enumerate()

````python
nombres = ["Ana", "Luis", "Pedro"]

for indice, nombre in enumerate(nombres):
    print(indice, nombre)
````

### Ejemplo 6: for con diccionarios

Podemos recorrer las claves y valores de un diccionario.

````python
precios = {"manzana": 100, "banana": 200, "cereza": 300}

for fruta, precio in precios.items():
    print(fruta, "cuesta", precio)
````

````python
for fruta in precios:
    print (fruta)
````

## 2.2 Bucle `while` en Python

El bucle `while` permite **ejecutar un bloque de código mientras se cumpla una condición**.  
Se utiliza cuando **no sabemos de antemano cuántas veces** queremos que se repita el ciclo.

---

### Sintaxis

```python
while condición:
    # código a ejecutar mientras la condición sea verdadera


### Ejemplo 1: Contador simple

````python
contador = 3

while contador <= 5:
    print("Número:", contador)
    contador += 1
````

### Ejemplo 2: while con condición booleana

````python
sigue = True
n = 2

while sigue:
    n += 1
    print("Vuelta", n)
    if n == 3:
        sigue = False
````

### Ejemplo 3: Bucle infinito con `break`

````python
while True:
    respuesta = input("Escribe 'salir' para terminar: ")
    if respuesta == "salir":
        print("Programa finalizado")
        break
````

### Ejemplo 4: Uso de `continue`

````python
numero = 2

while numero < 5:
    numero += 1
    if numero == 3:
        continue  # salta esta iteración
    print("Número:", numero)
````

````python
# Ejercicio 9: Dibujar triángulo centrado con for
altura = 27
for i in range(1, altura):
    espacios = " " * (altura - i)
    estrellas = "*" * (2 * i - 1)
    print(espacios + estrellas)
````
