# El ndarray <a name="ndarray"></a>
El ndarray es la columna vertebral de Numpy. Es un contenedor rápido y flexible para objetos de matriz N-dimensionales, generalmente utilizado para grandes conjuntos de datos en Python. Las matrices le permiten realizar operaciones matemáticas en bloques completos de datos utilizando una sintaxis similar a las operaciones equivalentes entre elementos escalares.

Aquí hay un ejemplo rápido de sus capacidades:

````python
# crear una matriz 2x3 de valores aleatorios
data = np.random.randn(2,3)
data
````

````python
data *10 #Multiplicación x escalar
````

Cada matriz tiene una forma, una tupla que indica el tamaño de cada dimensión y un dtype. Puede obtenerlos a través de los métodos respectivos:

````python
# el tamaño de la matriz

data.shape
````

````python
# número de dimensiones de la matriz

data.ndim
````

````python
# El tipo  almacenado en la matriz

data.dtype
````

## Data Types <a name="data"></a>
El tipo de datos o dtype es un objeto especial que contiene la información que la matriz necesita para interpretar un fragmento de memoria. Podemos especificarlo durante la creación de una matriz.

````python
arr = np.array([1, 2, 3], dtype=np.float64)
arr.dtype
````

````python
arr = np.array([1, 2, 3], dtype=np.int64)
arr.dtype
````

Un ndarray solo puede contener datos de **un** tipo de datos (dtype). Esto lo hace un poco menos flexible que una lista normal de Python, pero permite que NumPy se ejecute más rápido.


Se puede usar:
- `np.int64` para almacenar números enteros
- `np.float64` para almacenar números reales (con decimales)
- `np.bool` para almacenar valores `True` y `False` -> Deprecado

Al crear matrices en NumPy, el tipo se infiere (adivina), por lo que no necesita especificarlo explícitamente.

Similar a pandas, puede convertir (cast) una matriz de un dtype a otro usando el método `astype`:

````python
arr = np.array([1, 2, 3])
arr.dtype
````

````python
float_arr = arr.astype(np.float64)
float_arr.dtype
````

### Broadcasting

La aplicación de operaciones aritméticas a matrices de diferentes tamaños se llama **broadcasting**.

````python
# Ejemplo 1

A = np.ones((3,3))
A
````

````python
B = np.arange(3)
B
````

````python
A + B
````

````python
# Ejemplo 2

a = np.array([1, 2, 3])
b = 2
resultado = a + b
print(resultado)
````

¿qué pasó?

Estiramos 'B' para que coincida con la forma de 'A', el resultado es una matriz bidimensional

## Funciones universales 
NumPy proporciona funciones matemáticas familiares como sin, cos y exp. En NumPy, estas se llaman "funciones universales"*ufunc*. 

Dentro de NumPy, estas funciones operan por elementos en una matriz, produciendo una matriz como salida. Aquí hay un ejemplo de `sqrt` y` exp` ($e^x$):

````python
A = np.arange(10)
A
````

````python
np.sqrt(A)
````

````python
np.exp(A)
````

Otras funciones universales toman 2 matrices como entrada. Estas se llaman funciones *binarias*.

Por ejemplo, `maximum ()` selecciona los valores más grandes de dos matrices de entrada.

````python
x = np.random.randn(10)
x
````

````python
y = np.random.randn(10)
y
````

````python
np.maximum(x, y)
````

## Condicional
La función `np.where` es muy útil para realizar selecciones condicionales y se utiliza tanto en el contexto de arrays como en DataFrames y Series de pandas.

````python
a = np.array([1, 2, 3, 4, 5])
a
````

````python
# Entrega los Indices que cumplen con la(s) condiciones.
resultado = np.where(a > 3)
print (resultado)
````

````python
# Devuelve una matriz con el resultado 
resultado = np.where(a > 3, 'Mayor', 'Menor o igual')
print (resultado)
````

## Su uso en en Pandas

````python
import pandas as pd
import numpy as np

# Crear un DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})
df
````

````python
# Crear una nueva columna basada en una condición con np.where

df['C'] = np.where(df['A'] > 3, 'Mayor', 'Menor o igual')
print(df)
````

````python
# Crear una nueva columna basada en una condición con np.where con Multiples condiciones

df['D'] = np.where(     (df['A'] > 3) & (df['B'] < 50)    , 'Cumple', 'No cumple')
print(df)
````

````python
df['D'] = np.where((df['D']=="Cumple"), 'Bien', 'Mal')
print(df)
````

### Ventajas de np.where() en pandas y NumPy:
Permite realizar selecciones y asignaciones condicionales de manera vectorizada, lo que significa que es mucho más rápido que usar bucles.

````python
import pandas as pd
import numpy as np

# Crear un DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5]
})

df['C'] = np.where(    df['A'] > 3, 'Mayor', '')

df
````
