# Variables Numericas

````python
entero = 5
flotante = 1.6

print(entero)
print(flotante)
````

Se puede operar con los dos tipos de variables utilizando los operadores suma (+), resta (-), multiplicación (*) y división (/). Debemos tener en cuenta que el resultado de cualquier operación realizada con variables flotantes será otra variable flotante. El resultado de cualquier operación realizada con variables enteras, será otra variable entera, a excepción de la división, que será flotante.

## Operadores Matemáticos

| Operador | Descripción                   |
|----------|-------------------------------|
| `+`      | Suma                          |
| `-`      | Resta                         |
| `*`      | Multiplicación                |
| `/`      | División (float)              |
| `//`     | División entera (floor)       |
| `%`      | Módulo (residuo de división)  |
| `**`     | Exponenciación (potencia)     |

Recordar el orden de las operaciones: primero las potencias luego multiplicacion o division o módulo y finalmente suma o resta


````python
# Enteros
a = 10
b = 3
````

````python
# Suma
print(f"Suma: {a} + {b} = {a + b}")
````

````python
# Resta
print(f"Resta: {a} - {b} = {a - b}")
````

````python
# Multiplicación
print(f"Multiplicación: {a} * {b} = {a * b}")
````

````python
# División
print(f"División: {a} / {b} = {a / b}")
````

````python
# División entera
print(f"División entera: {a} // {b} = {a // b}")
````

````python
# Módulo
print(f"Módulo: {a} % {b} = {a % b}")
````

````python
# Exponenciación
print(f"Exponenciación: {a} ** {b} = {a ** b}")
````

````python
# Float
a = 10.0
b = 3.0

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b}")
print(f"División entera: {a} // {b} = {a // b}")
print(f"Módulo: {a} % {b} = {a % b}")
print(f"Exponenciación: {a} ** {b} = {a ** b}")
````

### También es posible mezclar variables que son tipo integer (enteras) y float (flotantes) en una operación. Sin embargo, notar que el resultado será de tipo float.

````python
# Float
a = 10
b = 3.0

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b}")
print(f"División entera: {a} // {b} = {a // b}")
print(f"Módulo: {a} % {b} = {a % b}")
print(f"Exponenciación: {a} ** {b} = {a ** b}")
````

# Operadores de Comparación

| Operador | Descripción                                         |
|----------|---------------------------------------------------|
| `==`     | Igual a                                           |
| `!=`     | Distinto de                                       |
| `>`      | Mayor que                                         |
| `<`      | Menor que                                         |
| `>=`     | Mayor o igual que                                 |
| `<=`     | Menor o igual que                                 |


````python
# Operadores de comparación
x = 5
y = 8

print(f"Igualdad: {x} == {y} -> {x == y}")
print(f"Desigualdad: {x} != {y} -> {x != y}")
print(f"Mayor que: {x} > {y} -> {x > y}")
print(f"Menor que: {x} < {y} -> {x < y}")
print(f"Mayor o igual: {x} >= {y} -> {x >= y}")
print(f"Menor o igual: {x} <= {y} -> {x <= y}")
````

# Conversion de tipos de variables

En Python, cada valor tiene un **tipo de dato** (`int`, `float`, `str`, `bool`, etc.).  
A veces necesitamos **convertir un valor de un tipo a otro**. Esto se conoce como **conversión de variables** o *type casting*.

### 1. Funciones básicas de conversión

| Función      | Descripción                                       |
|-------------|--------------------------------------------------|
| `int(x)`    | Convierte `x` a entero                           |
| `float(x)`  | Convierte `x` a número decimal (float)          |
| `str(x)`    | Convierte `x` a cadena (string)                 |
| `bool(x)`   | Convierte `x` a booleano (`True` o `False`)    |

### 2. Conversión a entero (`int()`)

- Convierte valores numéricos o strings que representan números enteros a `int`.  
- Redondea hacia cero si es un `float`.

````python
x = "42"
y = int(x)      # y = 42
z = int(3.9)    # z = 3
print (x,y,z)
````

### 3. Conversión a float (float())

Convierte valores numéricos o strings que representan números decimales a float.

````python
x = "3.14"
y = float(x)    # y = 3.14
z = float(5)    # z = 5.0
````

### 4. Conversión a string (str())

Convierte cualquier valor a string para poder imprimirlo o concatenarlo.

````python
x = 42
y = str(x)      # y = "42"
````

### 5. Conversión a booleano (bool())

Convierte valores a True o False según su "veracidad".

Valores considerados falsos: 0, 0.0, "" (string vacío), [] (lista vacía), {} (diccionario vacío), None.

````python
x = 0
y = bool(x)     # y = False

z = "Hola"
w = bool(z)     # w = True
````

### 6. Conversión entre tipos compatibles

Algunos tipos no se pueden convertir directamente (ej: "Hola" a int dará error).

Se debe asegurar que el valor sea compatible con el tipo destino.

````python
# Correcto
x = "123"
y = int(x)      # 123

# Incorrecto -> genera error (descomentar)
#x = "Hola"
#y = int(x)      # ValueError
````
