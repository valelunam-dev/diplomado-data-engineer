# 7. Agrupamiento (GroupBy)

```python
# Datos
data = {'Ciudad':['Santiago','Valparaíso','Santiago','Concepción','Valparaíso'],
        'Ventas':[100,200,150,300,120]}
df3 = pd.DataFrame(data)
df3
```

## 7.1 Agrupar por 1 columna

```python
# Agrupar y sumar
print(df3.groupby('Ciudad').sum())
```

```python
# Agrupar y promedio
print(df3.groupby('Ciudad').mean())
```

```python
# Agrupar y Contar registros
print(df3.groupby('Ciudad').count())
```

## 7.2 Agrupar por mas de 1 columna

```python
# Datos
data = {
    'Departamento': ['Ventas', 'Ventas', 'TI', 'TI', 'Ventas', 'TI'],
    'Ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Barcelona', 'Madrid', 'Barcelona'],
    'Ventas': [100, 150, 200, 250, 300, 350],
    'Empleados': [5, 3, 8, 6, 7, 4]
}

df = pd.DataFrame(data)
print(df)
```

### 7.2.1 Agrupar y Contar

```python
# Contar registros por Departamento y Ciudad
conteo = df.groupby(['Departamento', 'Ciudad']).size()
print(conteo)
```

```python
# Contar registros por Departamento y Ciudad
conteo = df.groupby(['Ciudad','Departamento']).size()
print (conteo)
```

### 7.2.2 Agrupar y Sumar

```python
# Agrupar por Departamento y Ciudad, y sumar las ventas
grupo = df.groupby(['Departamento', 'Ciudad'])['Ventas'].sum()
print(grupo)
```

### 7.2.3 Agrupar y aplicar múltiples operaciones

```python
# Agrupar por dos columnas y aplicar diferentes operaciones
resultado = df.groupby(['Departamento', 'Ciudad']).agg({
    'Ventas': ['sum', 'mean'],  # Suma y promedio de ventas
    'Empleados': 'mean'         # Promedio de empleados
})
print(resultado)
```

### 7.2.4 Agrupar con reset_index()

A menudo, es útil convertir el MultiIndex en columnas regulares. Se puede lograr esto encadenando el método .reset_index() al final.

```python
# Agrupar por Departamento y Ciudad, y sumar las ventas
grupo = df.groupby(['Departamento', 'Ciudad'])['Ventas'].sum().reset_index()
grupo
```

# 8. Merge y Concatenación

```python
# Datos
df1 = pd.DataFrame({'ID':[1,2,3], 'Nombre':['Ana','Juan','Pedro']})
df2 = pd.DataFrame({'ID':[2,3,4], 'Ciudad':['Valpo','Conce','Santi']})

print (df1)
print ("-------------")
print (df2)
```

```python
# Merge inner
print(pd.merge(df1, df2, on='ID', how='inner'))
```

```python
# Merge left
print(pd.merge(df1, df2, on='ID', how='left'))
```

```python
# Concatenar verticalmente
print(pd.concat([df1, df2], ignore_index=True))
```

# 9. Pivot

```python
# Datos
data = {
    'Producto': ['Manzana', 'Uva', 'Manzana', 'Uva', 'Palta', 'Palta', 'Manzana'],
    'Ciudad': ['Santiago', 'Santiago', 'Valparaíso', 'Valparaíso', 'Santiago', 'Concepción', 'Concepción'],
    'Ventas': [100, 150, 200, 250, 300, 350, 400],
    'Clientes': [5, 3, 8, 6, 7, 4, 9]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Productos como filas, Ciudades como columnas, Ventas como valores
pivot1 = df.pivot(index='Producto', columns='Ciudad', values='Ventas')
print("\n1. Ventas por Producto y Ciudad:")
print(pivot1)
```

```python
# Múltiples medidas (Ventas y Clientes)
pivot2 = df.pivot(index='Producto', columns='Ciudad', values=['Ventas', 'Clientes'])
print("\n2. Ventas y Clientes por Producto y Ciudad:")
print(pivot2)
```

```python
# pivot_table con múltiples funciones
pivot4 = df.pivot_table(
    index='Producto',
    columns='Ciudad',
    values='Ventas',
    aggfunc=['sum', 'mean', 'count'],  # Suma, promedio y conteo
    fill_value=0
)
print("\n4. Diferentes estadísticas por ciudad:")
pivot4
```

```python
# Para incluir todas las ciudades aunque no tengan datos
pivot5 = df.pivot_table(
    index='Producto',
    columns='Ciudad',
    values='Ventas',
    aggfunc='sum',
    fill_value=0
)
print("\n5. Ventas con todas las ciudades (fill_value=0):")
print(pivot5)
```

```python
import pandas as pd

# Datos con duplicados: Manzana-Santiago aparece 2 veces
data = {
    'Producto': ['Manzana', 'Manzana', 'Uva', 'Manzana'],
    'Ciudad': ['Santiago', 'Santiago', 'Valparaíso', 'Santiago'],
    'Ventas': [100, 50, 200, 75]
}

df = pd.DataFrame(data)
print("Datos con duplicados:")
print(df)
```

```python
# ERROR: No se puede pivotar porque hay valores duplicados
# df.pivot(index='Producto', columns='Ciudad', values='Ventas')
```

```python
# Suma los valores duplicados: 100 + 50 + 75 = 225 para Manzana-Santiago
pivot_sum = df.pivot_table(
    index='Producto', 
    columns='Ciudad', 
    values='Ventas', 
    aggfunc='sum'  # ← Suma los duplicados
)
print("Suma de ventas (agrega duplicados):")
print(pivot_sum)
```

Se puede buscar por  `pandas read_csv` en internet para buscar la documentación.

También hay muchas otras funciones que pueden leer datos. Algunos ejemplos:

| Función | Descripción
| -- | -- |

| read_table     | Cargue datos delimitados de un archivo, URL u objeto similar. El delimitador predeterminado es tab `\t` |

| read_hdf       | Leer archivos HDF5 escrito por pandas |
| read_html      | Leer todas las tablas encontradas en el documento HTML dado |
| read_json      | Leer datos de una cadena JSON |
| read_sql       | Leer los resultados de una consulta SQL |
