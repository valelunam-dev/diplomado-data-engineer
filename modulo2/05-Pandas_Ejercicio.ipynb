# 🎬 Ejercicio: Actores que trabajaron en Chile  

Una productora internacional ha contratado a diferentes actores de Hollywood para grabar una serie de películas en distintas ciudades de Chile.  
Cada registro en el dataset **`Horas_Trabajadas.csv`** corresponde a un actor trabajando en una ciudad y región de Chile, con las horas trabajadas durante el día y la noche.  

## 📂 Columnas del dataset
- **Nombre**: Nombre y apellido del actor.  
- **Ciudad**: Ciudad donde trabajó el actor.  
- **Region**: Región administrativa de Chile donde se ubica la ciudad.  
- **Horas Dia**: Número de horas trabajadas en el día.  
- **Horas Noche**: Número de horas trabajadas en la noche.  

La productora quiere **analizar la información para optimizar los tiempos de grabación y la distribución de trabajo**.
