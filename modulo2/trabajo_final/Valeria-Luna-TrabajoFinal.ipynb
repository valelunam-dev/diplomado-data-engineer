# Preguntas

## 1️⃣ Leer el archivo
- Importa Pandas y carga el archivo `Ventas.csv` en un DataFrame llamado `df`.
- Muestra las primeras 7 filas.

```python
import pandas as pd

df = pd.read_csv('Ventas.csv', sep=",", encoding='utf-8') #separado por comas
df.head(7)
```

## 2️⃣ Inspección

- Cuantas filas y columnas tiene el Dataframe.

```python
(df.shape)
```

## 3️⃣ Selección de columnas

- Muestra solo las columnas Cliente y Producto.

```python
df[['Cliente', 'Producto']]
```

## 4️⃣ Filtrado de datos

- Muestra todas las ventas que se hicieron en la zona "Norte".

```python
df[df['Zona']=='Norte'] #solo zona norte
```

## 5️⃣ Operaciones Básicas

- Crea una nueva columna Total con el cálculo de (Cantidad * Precio).
- Muestra el DataFrame actualizado ( 8 primeras filas)

```python
df['Total'] = df['Cantidad'] * df['Precio'] #crea nueva columna con operacion en el df
df.head(8)
```

## 6️⃣ Análisis

- ¿Cuál fué el total de ventas del día?
- ¿Cuál es la venta por cada zona?

```python
# Total de ventas del día
total_ventas = df['Total'].sum() #Sumamos la nueva columna total para obtener el total de ventas del día
print('Total de ventas del día:', total_ventas)

# Venta por cada zona
ventas_por_zona = df.groupby('Zona')['Total'].sum() #Agrupamos por zona y sumamos el total de ventas
print('\nVentas por zona:')
print(ventas_por_zona)
```

## 7️⃣ Guardar Resultados

- Generar un archivo "resumen.csv" que contenga:

    - Suma de Ventas por Zona (Descripcion y Total)
    - Suma de Ventas por producto (Descripcion y Total)
    - Cantidad de registros (transacciones) : xxxxx
    - Suma de Ventas: yyyyy

- El archivo debiera tener 3 columnas:
    - Descripcion 
    - Valor
    - Tipo

- Ejemplo:


| Descripción | Valor  | Tipo      |
|------------|-----------|-----------|
| Centro     | 711,993   | Zonas     |
...
| Sur        | 624,513   | Zonas     |
| Banana     | 429,714   | Productos |
| Cereza     | 107,314   | Productos |
...
| Piña       | 335,352   | Productos |
| Uva        | 352,107   | Productos |
| venta      | 2,702,306 | Resumen   |
| registros  | 1,000     | Resumen   |



Ayuda: Algunas maneras de renombrar Las columnas de un dataframe:

- df = df.rename(columns={'nombre_antiguo': 'nombre_nuevo'})
- df.rename(columns={"A": "Columna1", "B": "Columna2"}, inplace=True)
- df.columns = ['nuevo_nombre1', 'nuevo_nombre2', ...]

```python
# Suma de ventas por zona
ventas_por_zona = df.groupby('Zona')['Total'].sum().reset_index()
ventas_por_zona = ventas_por_zona.rename(columns={'Zona': 'Descripcion', 'Total': 'Valor'})
ventas_por_zona['Tipo'] = 'Zonas'
ventas_por_zona
```

```python
# Suma de ventas por producto
ventas_por_producto = df.groupby('Producto')['Total'].sum().reset_index()
ventas_por_producto = ventas_por_producto.rename(columns={'Producto': 'Descripcion', 'Total': 'Valor'})
ventas_por_producto['Tipo'] = 'Productos'
ventas_por_producto
```

```python
# Resumen general
total_ventas = df['Total'].sum()
registro_count = len(df)
resumen_general = pd.DataFrame([
    {'Descripcion': 'Venta', 'Valor': total_ventas, 'Tipo': 'Resumen'},
    {'Descripcion': 'Registros', 'Valor': registro_count, 'Tipo': 'Resumen'}
])
resumen_general
```

```python
# Combinar en un solo DataFrame
df_resumen = pd.concat([ventas_por_zona, ventas_por_producto, resumen_general], ignore_index=True)

#Guardar el resumen en un archivo CSV y mostrar el nuevo df_resumen
df_resumen.to_csv('Valeria-Luna-resumen.csv', index=False, sep=';',encoding='utf-8')
df_resumen
```
