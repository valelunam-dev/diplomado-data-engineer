## Estructuras de datos incorporadas
Hemos visto tipos simples de Python: int, float, bool y str. Python también tiene varios tipos compuestos integrados, que actuan como contenedores para otros tipos. Estos tipos compuestos son:

| Tipo  | Ejemplo | Descripción |
|----|---|---|
| list  | [1,2,3] | Colección ordenada |
| tupla  | (1,2,3) | Colección ordenada, inmutable |
| dict| {'a':1, 'b':2, 'c':3} | Mapeo desordenado (clave/key, valor) |
| set  | {1,2,3} | Colección desordenada de valores únicos |

# 1. Listas

En Python, una **lista** es una colección **ordenada y modificable** de elementos.  
Se define usando corchetes `[]` y puede contener números, cadenas, booleanos u otros objetos.

---
### Crear una lista

```python
frutas = ["manzana", "banana", "cereza"]
numeros = [10, 20, 30, 40]
mixta = [1, "hola", True, 3.14]

## Acceder a los elementos

````python
frutas = ["manzana", "banana", "cereza"]

print(frutas[0])   # primer elemento -> manzana
print(frutas[1])   # segundo elemento -> banana
print(frutas[-1])  # último elemento -> cereza
print(frutas[-2]) 
````

## Modificar elementos

````python
frutas = ["manzana", "banana", "cereza"]

frutas[1] = "pera"
print(frutas)  # ["manzana", "pera", "cereza"]
````

## Recorrer una lista con for

````python
numeros = [10, 20, 30]

for n in numeros:
    print("Número:", n)
````

## Operaciones con listas

````python
numeros = [1, 2, 3]

print(len(numeros))      # longitud -> 3
print(sum(numeros))      # suma -> 6
print(max(numeros))      # máximo -> 3
print(min(numeros))      # mínimo -> 1
````

## Métodos útiles de listas

Los **métodos más utilizados** con las listas son los siguientes:

| Método | Acción |
|--|--|
| `alumnos.append("Javiera")` | Inserta "Javiera" al final de la lista |
| `alumnos.insert(0,"Javiera")` | Inserta "Javiera" en la posición 0 |
| `alumnos.remove("Javiera")` | Elimina la primera aparición de "Javiera" de la lista |
| `alumnos.pop()` | Elimina el último elemento de la lista |
| `alumnos.pop(3)` | Elimina el cuarto elemento de la lista |
| `alumnos.clear()` | Elimina todos los elementos de la lista |
| `alumnos.index("Javiera")` | Devuelve el índice de la primera aparición de "Javiera" |
| `alumnos.sort()` | Ordena la lista (los elementos deben ser comparables) |
| `sorted(alumnos)` | Devuelve una copia de la lista 'alumnos' ordenada   |
| `alumnos.reverse()` | Ordena la lista en orden inverso |
| `alumnos.copy()` | Devuelve una copia de la lista |
| `alumnos.extend(otra_lista)` | Fusiona las dos listas |

````python
frutas = ["manzana", "banana", "cereza"]

frutas.append("pera")       # agrega al final
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.insert(1, "uva")     # inserta en posición 1
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.remove("banana")     # elimina un elemento
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.pop()                # elimina el último
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.pop(1)                # elimina por indice
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.clear()                # elimina contenido
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

indice = frutas.index("banana")  # indica indice
print(indice)
````

````python
frutas = ["manzana", "banana", "cereza"]

frutas.sort()               # ordena alfabéticamente
print(frutas)
````

````python
frutas = ["manzana", "banana", "cereza"]

a = frutas.reverse()           # invierte el orden
a
````

## Rebanado (slicing)

````python
numeros = [10, 20, 30, 40, 50]

print(numeros[1:4])   # [20, 30, 40] 
print(numeros[:3])    # [10, 20, 30]
print(numeros[2:])    # [30, 40, 50]
print(numeros[::2])   # [10, 30, 50]
````

## Listas anidadas

````python
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(matriz[0])      # [1, 2, 3]
print(matriz[1][2])   # 6
````

# 2. Tuplas

En Python, una **tupla** es una colección **ordenada e inmutable** de elementos.  
Son similares a las listas, pero **no se pueden modificar** (no se pueden agregar, eliminar o cambiar elementos).  

Se definen usando **paréntesis `()`** en lugar de corchetes `[]`.

---

### Crear una tupla

```python
frutas = ("manzana", "banana", "cereza")
numeros = (10, 20, 30, 40)
mixta = (1, "hola", True, 3.14)

## Acceder a los elementos

````python
frutas = ("manzana", "banana", "cereza")

print(frutas[0])   # primer elemento -> manzana
print(frutas[1])   # segundo elemento -> banana
print(frutas[-1])  # último elemento -> cereza
````

## Recorrer una tupla con for

````python
numeros = (10, 20, 30)

for n in numeros:
    print("Número:", n)
````

## Operaciones con tuplas

````python
numeros = (1, 2, 3, 2, 4, 2)

print(len(numeros))       # longitud -> 6
print(numeros.count(2))   # cuántas veces aparece 2 -> 3
print(numeros.index(4))   # índice donde aparece 4 -> 4
````

## Tuplas anidadas

````python
tupla_anidada = ((1, 2), (3, 4), (5, 6))

print(tupla_anidada[0])     # (1, 2)
print(tupla_anidada[1][1])  # 4
````

# 3. Sets

En Python, un **set (conjunto)** es una colección **desordenada, sin índices y sin elementos duplicados**.  
Se definen usando llaves `{}` o con la función `set()`.

---

## Crear un set

```python
frutas = {"manzana", "banana", "cereza"}
numeros = {1, 2, 3, 4, 5}
vacio = set()   # conjunto vacío

## Creación de Set

Se dice que un Set es "No ordenado", ya que no queda registrado el orden en que se ingresan los elementos

````python
numeros = {2, 3, 4, 5, 1}
print (numeros)
````

## Comprobar si un elemento está en el set

````python
frutas = {"manzana", "banana", "cereza"}

print("banana" in frutas)   # True
print("pera" in frutas)     # False
````

## Agregar y eliminar elementos

````python
frutas = {"manzana", "banana"}

frutas.add("cereza")        # agrega un elemento
frutas.remove("banana")     # elimina un elemento (error si no existe)
frutas.discard("pera")      # elimina sin error si no existe
frutas.clear()              # elimina todos los elementos
````

## Operaciones de conjuntos

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # unión -> {1, 2, 3, 4, 5, 6}
print(A & B)   # intersección -> {3, 4}
print(A - B)   # diferencia -> {1, 2}
````

## Convertir lista en set (para eliminar duplicados)

````python
lista = [1, 2, 2, 3, 4, 4, 5]
conjunto = set(lista)

print(conjunto)  # {1, 2, 3, 4, 5}
````

# 4. Diccionarios

En Python, un **diccionario (`dict`)** es una colección **desordenada y modificable** de elementos, que se almacena en **pares clave:valor**.

Se definen usando llaves `{}`.

---

### Crear un diccionario

```python
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Santiago"
}

## Acceder a valores

````python
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Santiago"
}

print(persona["nombre"])     # Ana
print(persona["edad"])       # 25 

# Usando get() (no da error si la clave no existe)
# print(persona.get("ciudad")) # Santiago
print(persona.get("pais","NE"))
````

## Modificar valores

````python
persona = {"nombre": "Ana","edad": 25,"ciudad": "Santiago"}

persona["edad"] = 26
persona["ciudad"] = "Valparaíso"
print(persona)
````

## Agregar y eliminar elementos

````python
persona = {"nombre": "Ana","edad": 25,"ciudad": "Santiago"}

persona["profesion"] = "Ingeniera"   # agregar
print(persona)

persona.pop("edad")                  # eliminar por clave
print(persona)

del persona["ciudad"]                # eliminar con del
print(persona)

persona.clear()                      # eliminar todo
print(persona)
````

## Recorrer un diccionario

````python
estudiante = {"nombre": "Luis", "edad": 20, "carrera": "Informática"}

# Solo claves
for aaa in estudiante:
    print(aaa)
print ("--------------------------")
# Claves y valores
for ccc, valor in estudiante.items():
    print(ccc, ":", valor)
````

## Métodos útiles

````python
estudiante = {"nombre": "Luis", "edad": 20, "carrera": "Informática"}

print(estudiante.keys())     # todas las claves
print(estudiante.values())   # todos los valores
print(estudiante.items())    # pares clave:valor
````

## Diccionarios anidados

````python
alumnos = {
    "A001": {"nombre": "Ana", "nota": 90},
    "A002": {"nombre": "Pedro", "nota": 75},
    "A003": {"nombre": "Lucía", "nota": 60}
}

print(alumnos["A001"]["nombre"])  # Ana
print(alumnos["A002"]["nota"])    # 75
````

## Diccionario con listas

````python
cursos = {
    "Matemáticas":{90, 85, 88},
    "Historia": [{'a','b'}, 80, 95]
}

print(cursos["Matemáticas"])      # [90, 85, 88]
print((cursos["Historia"]))    # 245
````

# 📊 Resumen de Estructuras de Datos en Python

| Estructura | Sintaxis | Características | Ejemplos de uso | Métodos principales |
|------------|----------|-----------------|-----------------|---------------------|
| **Lista** (`list`) | `[1, 2, 3]` | - Ordenada<br>- Mutable (se puede modificar)<br>- Admite duplicados | Lista de alumnos, carrito de compras | `.append()`, `.remove()`, `.pop()`, `.sort()`, `.reverse()`, `.extend()` |
| **Tupla** (`tuple`) | `(1, 2, 3)` | - Ordenada<br>- Inmutable (no se modifica)<br>- Admite duplicados | Coordenadas (x, y), fechas, datos que no deben cambiar | `.count()`, `.index()` |
| **Diccionario** (`dict`) | `{"clave": "valor"}` | - No ordenado (desde 3.7 mantiene orden de inserción)<br>- Claves únicas<br>- Mutable | Agenda de contactos, inventario de productos | `.keys()`, `.values()`, `.items()`, `.get()`, `.update()`, `.pop()` |
| **Set** (`set`) | `{1, 2, 3}` | - No ordenado<br>- No permite duplicados<br>- Mutable | Conjunto de hobbies, elementos únicos, eliminar repetidos | `.add()`, `.remove()`, `.discard()`, `.union()`, `.intersection()`, `.difference()` |
