```python
import pandas as pd
import json
import requests
url = "https://randomuser.me/api/?results=5"
respuesta = requests.get(url)
datos_json = respuesta.json()
print(json.dumps(datos_json['results'][0]['location'],indent=2))
print(json.dumps(datos_json['results'],indent=2))
```

```python
import pandas as pd
df_usuarios = pd.json_normalize(datos_json['results'])
df_limpio = df_usuarios[['gender', 'email', 'name.first','name.last', 'location.country']]
df_limpio = df_limpio.rename(columns={'location.country':'pais'})
df_limpio.head()
```

```python
import pandas as pd
df_bonito = pd.json_normalize(
    df_usuarios,
        record_path=['results'],
        meta=['gender', ['location', 'street', 'name']]
)
print(df_bonito)
```
