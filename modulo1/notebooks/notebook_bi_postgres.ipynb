## 📊 Consulta SQL

```python
query = '''
SELECT 
    c.name as categoria,
    COUNT(*) as total_peliculas
FROM category c
JOIN film_category fc ON c.category_id = fc.category_id
GROUP BY c.name
ORDER BY total_peliculas DESC
'''

df = pd.read_sql(query, conn)
df.head()
```

## 🔄 Transformación de datos

```python
# Ejemplo: calcular porcentaje
df['porcentaje'] = df['total_peliculas'] / df['total_peliculas'].sum()
df
```

## 📈 Visualización

```python
plt.figure()
plt.bar(df['categoria'], df['total_peliculas'])
plt.xticks(rotation=45)
plt.title('Películas por categoría')
plt.tight_layout()
plt.show()
```
