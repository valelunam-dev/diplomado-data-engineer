Usted es trabajador de uber, y le han pasado el siguiente dataset viajes de Uber en 2024. Se le ha proporionado la función `readFile` que lee el archivo 'ncr_ride_bookings.csv' y le entrega una lista con los nombres de las columnas, y una lista con cada una de las filas del dataset, tal como lo puede ver en el código de a continuación.


```python
import csv

def readFile(fileName='uber.csv'):
    with open(fileName, mode='r') as file:
        reader = csv.reader(file)
        columnsName = next(reader) # Leer la primera fila como encabezado
        data = [row for row in reader]  # Leer el resto de las filas como datos
        return columnsName, data
    
columnsName, data = readFile()
print("columnsName:\n", columnsName)
print("First 5 rows of data:")
for i, row in enumerate(data[:5]):
    print(i, row)
```

## Desarrollo de la Actividad
Implemente las siguientes funciones:
1. `cleanDataset`: el cual limpia el conjunto de datos, eliminando todas filas que cumplas alguna de las siguientes condiciones:
    * Como estado distinto a "Completed" (es decir, debe dejar solo las filas con estado "Completed").
    * Que tengan "Booking Value", "Ride Distance" o "Driver Ratings" en "null".
    Puede añadir los parámetros que considere necesarios.

La función debe retornar una lista con los datos limpios (puede ser una lista nueva).\
El tiempo de ejecución de su función debe ser $O(n)$, donde $n$ es la cantidad de filas del dataset.

2. Implemente la función `avgColumns`, que retorne una lista con el promedio de las columnas: "Booking Value", "Ride Distance" y "Driver Ratings", de los datos ya limpios.\
*Nota*: debe retornar los datos en la lista en el orden descrito anteriormente.

```python
#1.cleanDataset:

def cleanDataset(columns, data):
    # Obtener índices de las columnas importantes
    status_idx = columns.index("Booking Status")
    booking_idx = columns.index("Booking Value")
    distance_idx = columns.index("Ride Distance")
    rating_idx = columns.index("Driver Ratings")
    
    clean_dataset = []
    
    for row in data:  # O(n)
        if (
            row[status_idx] == "Completed" and
            row[booking_idx] != "null" and
            row[distance_idx] != "null" and
            row[rating_idx] != "null"
        ):
            clean_dataset.append(row)
    
    return clean_dataset

data_clean = cleanDataset(columnsName, data)
print(columnsName)
for i, row in enumerate(data_clean[:20]):
    print(i, row)

#import pandas as pd # si quiero convertir el resultado a un dataframe para analizarlo mejor
#df_clean = pd.DataFrame(data_clean, columns=columnsName)
#df_clean
#print(df_clean.info())    #si quiero ver cuantos nonulos hay en cada columna
#print((df_clean["Booking Status"] != "Completed").all()) #para verificar que no hay filas con estado distinto a "Completed"
```

```python
#2.avgColumns:

def avgColumns(columns, data):
    booking_idx = columns.index("Booking Value")
    distance_idx = columns.index("Ride Distance")
    rating_idx = columns.index("Driver Ratings")
    
    sum_booking = 0
    sum_distance = 0
    sum_rating = 0
    
    n = len(data)
    
    for row in data:  # O(n)
        sum_booking += float(row[booking_idx])
        sum_distance += float(row[distance_idx])
        sum_rating += float(row[rating_idx])
    
    return [
        round(sum_booking / n, 2),
        round(sum_distance / n, 2),
        round(sum_rating / n, 2)
    ]

data2 = cleanDataset(columnsName, data)
promedios = avgColumns(columnsName, data2)
print(promedios)

```
