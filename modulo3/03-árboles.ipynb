# Árboles Binarios

Desafortunadamente en Python no existe una implementación nativa en la librería estándar.
Por lo tanto, necesitaremos la siguiente definición de nodo para poder representar un árbol binario.
Ojo esto es por simplicidad para el módulo en la practica podría perfectamente utilizar alguna librería que le permita usar árboles.


```python
class Node:
    def __init__(self, key, value=None):
        self.key = key
        self.value = value
        self.left = None
        self.right = None
```

No es necesario que entienda el código siguiente, lo utilizaremos para mostrar la estructura del árbol.

```python
def printBinaryTree(root):
    """Función principal que inicia la impresión del árbol con arcos."""
    if root:
        _recursivePrint(root, "", True)

def _recursivePrint(node, prefix, isLast):
    """
    Función recursiva auxiliar para imprimir el árbol.
    Utiliza un recorrido in-order inverso (Derecha -> Raíz -> Izquierda).
    """
    if node is not None:
        # Imprime la línea del conector para este nodo
        # Si es el último hijo (o el de la izquierda), usa '└──', si no, '├──'
        print(prefix + ("└── " if isLast else "├── ") + str(node.key))
        
        # Prepara los prefijos para los nodos hijos
        # Si este nodo era el último, el prefijo para sus hijos no necesita la línea vertical '|'
        childPrefix = prefix + ("    " if isLast else "│   ")
        
        # Recopila los hijos para saber cuál es el último
        children = [child for child in [node.left, node.right] if child is not None]
        
        for i, child in enumerate(children):
            # Llama recursivamente, marcando si el hijo es el último de la lista
            isLastChild = (i == len(children) - 1)
            _recursivePrint(child, childPrefix, isLastChild)
```

### Representemos un árbol de decisión binario

Este árbol de ejemplo nos permitirá tomar una sobre la decisión si comer o no?, solo tomando decisiones del estilo sí o no.

```python
root = Node("Comer?")
root.left = Node("Salir")
root.right = Node("Almuerzo")
root.right.left = Node("Pollo Frito")
root.right.left.left = Node("Con Papas")
root.right.left.right = Node("Con Arroz")
root.right.right = Node("Ensalada")
root.right.right.left = Node("Con Tomate")
root.right.right.right = Node("Con Lechuga")

# BFS(root)
printBinaryTree(root)
```

### Recorridos sobre árboles binarios

```python
from queue import Queue
def BFS(root):
    if root is None:
        return
    
    q = Queue()
    q.put(root)

    while not q.empty():
        nodesPerLevel = q.qsize()
        for _ in range(nodesPerLevel):
            current = q.get()
            print(current.key, end=' ')
            if current.left is not None:
                q.put(current.left)   
            if current.right is not None:
                q.put(current.right)
        print()

def _recursivePreorder(node, traverse):
    if node is None:
        return
    traverse.append(node)
    _recursivePreorder(node.left, traverse)
    _recursivePreorder(node.right, traverse)

def _recursiveInorder(node, traverse):
    if node is None:
        return
    _recursiveInorder(node.left, traverse)
    traverse.append(node)
    _recursiveInorder(node.right, traverse)

def _recursivePostorder(node, traverse):
    if node is None:
        return
    _recursivePostorder(node.left, traverse)
    _recursivePostorder(node.right, traverse)
    traverse.append(node)
    
def DFS(root, type="preorder"):
    traverse = []
    if type == "preorder":
        _recursivePreorder(root, traverse)
    elif type == "inorder":
        _recursiveInorder(root, traverse)
    else:
        _recursivePostorder(root, traverse)
    return traverse
```

## Un Ejemplo de Recorrido

```python
rootAE = Node("+")

rootAE.left = Node("x")
rootAE.left.left = Node(2)
rootAE.left.right = Node("-")
rootAE.left.right.left = Node(10)
rootAE.left.right.right = Node(5)

rootAE.right = Node("x")
rootAE.right.left = Node (3)
rootAE.right.right = Node (-1)

printBinaryTree(rootAE)

postorderTraversal = DFS(rootAE, "postorder")
[node.key for node in postorderTraversal]
```

## Binary Search Trees (BST)

```python
# Estructura del BST
class BinarySearchTree:
    def __init__(self):
        self.root = None
        self.height = 0
        self.length = 0

    def _insertRecursively(self, node, newNode, currentHeight):
        if newNode.key < node.key:
            if node.left is None:
                node.left = newNode
                self.length += 1
                if currentHeight + 1 > self.height:
                    self.height += 1
                return True
            else:
                return self._insertRecursively(node.left, newNode, currentHeight + 1)
        if newNode.key > node.key:
            if node.right is None:
                node.right = newNode
                self.length += 1
                if currentHeight + 1 > self.height:
                    self.height +=1
            else:
                return self._insertRecursively(node.right, newNode, currentHeight+1)
        else:
            return False
    
    def insert(self, key, value=None):
        newNode = Node(key, value)
        if self.root is None:
            self.root = newNode
            self.height = 0
            self.length = 1
        else:
            return self._insertRecursively(self.root, newNode, 0)
        
L = [1, 10, 14, 6, 5, 7, 0, 13]
bst = BinarySearchTree()
for x in L:
    bst.insert(x)
printBinaryTree(bst.root)
[node.key for node in DFS(bst.root)]
```
