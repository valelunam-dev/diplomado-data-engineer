```python
from sys import maxsize # número muy grande

def sort(lista):
    n = len(lista) # largo de la lista
    listaOrdenada = []
    for i in range(0, n): # de i=0 a i=n-1
        min = maxsize
        minIndex = 0
        for j in range(0, n): # de j=0 a j=n-1
            if lista[j] != "x" and lista[j] < min:
                min = lista[j]
                minIndex = j
        lista[minIndex] = "x"
        listaOrdenada.append(min)
    
    return listaOrdenada
```

```python
# Ejemplo de utilización de la función sort
lista = [2, 5, 1, 4, 6, 10, 3]
print(sort(lista))
```

```python
1. Explique brevemente qué hace el algoritmo.

El algoritmo implementa una estrategia de selección repetida del mínimo:
- Recorre la lista completa buscando el valor mínimo disponible.
- Una vez encontrado lo agrega a una nueva lista (listaOrdenada), marca ese elemento como usado con una "x" en la lista original.
- Repite este proceso n veces, hasta que todos los elementos han sido seleccionados.
```

```python
2. Cuál es el tiempo de ejecución del algoritmo?
* Puede asumir que el operador `append()` toma tiempo  $\mathcal{O(1)}$.

from sys import maxsize # número muy grande

def sort(lista):
    n = len(lista) #1
    listaOrdenada = [] #1
    #-----------------------------------------------------------------------n(5n+4)
    for i in range(0, n): #n
        min = maxsize #1
        minIndex = 0 #1
        #----------------------------------------------------------5n
        for j in range(0, n): #n
            if lista[j] != "x" and lista[j] < min: #3
                min = lista[j] #1
                minIndex = j #1
        #----------------------------------------------------------        
        lista[minIndex] = "x" #1
        listaOrdenada.append(min) #1
    #-----------------------------------------------------------------------
    return listaOrdenada #1

T(n) = 5n^2 + 4n + 3
```
