# Estructuras de Datos Lineales

Las estructuras de datos lineales son aquellas que para cada elemento de la estructura de datos, tienen un único un antecesor y un sucesor, exceptuando el primer y último elemento.

Este tipo de estructuras nos serán de utilidad cuando necesitemos manipular datos siguiendo un orden particular, y no búscar sobre los mismos.
Por ejemplo:
* Modelar el orden cronológico en que ocurren eventos.
    * Solicitudes de impresión a una impresora
    * Procesos que deben ejecutarse en un sistema operativo(windows, linux, macos, etc)
* Procesar datos en orden cronológico inverso.
    * Operación "undo" en editores de texto (ctrl+z)
    * Función "volver" en un navegador web
    * etc

A continuación veremos las estructuras de datos lineales más importantes:
* `Listas`,
* `Pilas`, y
* `Colas`



## 1. Arreglos
Antes de ver las estructuras de datos antes mencionadas veamos una estructura de datos primitiva muy importante. Los `arreglos` nos será útil para entender las demás.

Los arreglos son una estructura de datos primitiva, por decir la más básica de todas que nos permite construir estructuras de datos más complejas.
Los arreglos permiten almacenar un conjunto datos en memoria principal (RAM: Random Access Memory), y tiene las siguientes propiedades muy importantes:
* Todos los datos tienen el mismo tipo (`int`, `float`, `string`, etc), se conoce como el *tipo base* del arreglo.
* Los elementos del arreglo *se almacenan de forma contigua detro de la memoria*, esta es quirás la propiedad más importante.
* Los elementos se acceden usando un *ídice*: primer elemento, segundo elemento, ...

Algo así seria un diagrama simplificado

```
+-----------------------------------+
|            MEMORIA RAM            |
|       (Direcciones de Memoria)    |
+-----------------------------------+
|               ...                 |
|               ...                 |
|               ...                 |
|-----------------------------------|
|         [ Elemento 0 ]            |   <-- Inicio del arreglo
|-----------------------------------|
|         [ Elemento 1 ]            |
|-----------------------------------|
|         [ Elemento 2 ]            |
|-----------------------------------|
|         [ Elemento 3 ]            |
|-----------------------------------|
|               ...                 |
|-----------------------------------|
|         [ Elemento n-1 ]          |
|-----------------------------------|
|               ...                 |
|               ...                 |
|               ...                 |
+-----------------------------------+
```

## 2. Listas
Las listas probablemente ya las conocen de nombre en Python.\
Una lista es una secuencia finita y *ordenada* de datos, donde:
* Cada elemento es llamado "elemento".
* Se dicen que son "*ordenadas*", en el sentido que cada elemento tiene asociado una posición dentro de la lista.
    * Primer, segundo, tercer elemento ...
* Dependiendo de la implementación, los datos que almacene pueden ser del mismo o distinto tipo.
    * En python pueden ser de distinto tipo por ejemplo.
Podemos resumir la definición como:
$L = \langle e_0, e_1, \ldots, e_{n-1}\rangle$

Finalmente, una carterística que es quizás la más importante es que son dinámicas:
* Pueden variar su tamaño a lo largo de la ejecución de nuestro programa.

### 2.1. Listas en Python
Las listas en python están implementadas sobre arreglos, lo que quiere decir que todos los datos están almacenados de forma contigua en memoria,
una característica muy importante de aquí en adelante, ya que implicará tenerlo en cuenta al momento de usar sus operadores.

Antes de cualquier cosa veamos como utilizarlas:

#### 2.1.1 Crear listas

Para crearlas es bastante sencillo, podemos utilizar corchetes `[]` ó `list()`

````python
# Lista vacía
l = []
l = list()
# También se puede inicializar con elementos
l = [1, 2, 3, 4, 5,"hola", 2.3, ["2"]] # note que también pueden ser de distintos tipos
l = list((1, 2, 3, 4, 5, "hola"))
print(l)
````

### 2.1.3 Operaciones no Mutadoras

#### Largo de una Lista
Para saber la cantidad de elementos que tiene una lista podemos usar el operador `len()`, generalmente a este dato se le conoce como el largo de la lista:

````python
length = len(l)
print("Largo de la lista:", length)
````

#### Acceso a elementos
* Podemos acceder por su posición: `L[i]`
* Podemos saber el índice en que está la primera aparición de un elemento: `L.index(x)`
* Podemos contar la cantidad de veces que un elemento `x` está presente: `L.count(x)`
* Podemos acceder a sublistas de `L`: `L[i:j]`, donde accedemos a los elementos entre los indices `i` y `j-1`.
* También podemos preguntar si un elemento está presente en la lista: `x in L`


````python
L = [1, 2, 3, 4, 5]
L[2:3]
````

````python
### Operador L[i]
# posiciones:
#    0, 1, 2, 3, 4, 5,  6,  7
L = [1, 2, 3, 4, 6, 5, 10, 4]
print("-----L[i]-----")
print("Elemento en la posición 4, L[4]:", L[4]) # Acceso al elemento en la posición 4

### Operador L.index(x)
print("-----L.index(x)-----")
print("Posición de la primera aparición del elemento 4, L.index(4):", L.index(4))
print("Posición de la primera aparición del elemento 6, L.index(6):", L.index(6))

### Operador L.count(x)
print("-----L.count(x)-----")
print("Cantidad de veces que aparece el elemento 4 en L, L.count(4):", L.count(4))
print("Cantidad de veces que aparece el elemento 1 en L, L.count(1):", L.count(1))
print("Cantidad de veces que aparece el elemento 11 en L, L.count(11):", L.count(11))

### Operador in
print("----- x in L -----")
print("¿El elemento 11 está en L?:", 11 in L)
print("El elemento 1 está en la lista?:", 1 in L)
````

#### Comparar entre Listas

* Podemos hacer comparaciones de igualdad: `Lista1 == Lista2`
    * Este operador retorna `True` si todos los elmentos son idénticos de ambas listas y además tienen el mismo orden.
*  Podemos hacer comparaciones del estilo `<, >, <=, >=`:
    * Este tipo de operadores compará elemento a elemento hasta que se cumpla para algún elemento.

````python
L1 = [2, 3, 4, 5, 6, 7]
L2 = [2, 3, 4, 5, 6, 7]
L3 = [2, 3, 4, 5, 6, 8]
print("L1 == L2:", L1 == L2)
print("L1 <= L3:", L1 <= L3)
L1[-1] = 8
print("L1 < L3:", L1 < L3)
print("L1 >= L3:", L1 >= L3)
L1[-1] = 9
print("L1 > L3:", L1 > L3)
````

#### Crear Nuevas Listas
* Podemos concatenar dos listas con el operador `+`, creando una nueva lista.
    * Se preserva el orden de las listas tal cual se llaman con el operador, es decir:\
    `L1+L2`, primero van los elementos de `L1`y luego los de `L2` en la nueva lista. 
* Podemos repetir `c`veces el contenido de una lista, creando una nueva lista.

````python
newL = L1 + L3
print("Concatenación de L1 y L3:", newL)
repeatL1 = 3*L1
print("repetición de L1 3 veces:", repeatL1)
````

### Costo de todas estas operaciones

|c * L|	O(cn)|	Hay que replicar la lista L, c veces en una nueva lista

### 2.1.4 Operadores Mutadores (pueden cambiar el contenido de la lista)
#### Inserción
* `L[i] = value`: podemos acceder a una casilla y asignarle un valor
* `L.append(value)`: podemos agregar elementos al final de la lista
* `L.insert(i, value)`: podemos agregar un elemento en una posición arbitraria `i`

#### Eliminación
* `L.pop()`: eliminamos el último elemento de L.
    * También podemos eliminar un elemento en alguna posición arbitraria `i` con:
        * `L.pop(i)` ó `del L[i]`
* `L.remove(value)`: eliminamos la primera ocurrencia de `value` en `L`

#### Otros operadores
* `L1.extend(L2)`: Podemos extender una lista, es similar al operador `+`, pero ojo no crea una nueva lista, exitiende sobre la misma lista `L1`.
    * Este comportamiento también se puede hacer con `L1+=L2`\
* `L.reverse()`: invierte la lista.

#### Costo de los operadores
|Operación|	Tiempo de Ejecución| Descripción|
|---------|--------------------|------------|
|L[i] = val|	O(1)|	-
|L.append(value)|	O(1)|	-
|L.insert(i, value)|	O(n-i+1)|	Debe desplazar los n-i valores de la lista a la derecha.
|L.pop()|	O(1)|	Solo debe borrar el último elemento
|L.pop(i) ó del L[i]|	O(n-i)|	Debe desplazar los n-i valores a izquierda de la lista.
|L.remove(value)|	O(n)|	En el peor de los casos debe recorrer toda la lista
|L1.extend(L2) ó L1 += L2|	O(n2)|	Agrega los n2 elementos de L2 en L1
|L.reverse()|	O(n)|	Debe recorrer toda la lista para invertiral
|L.sort()|	O(n log2 n)|	Usa una variante de QuickSort (solo como dato)

## 3. Pilas

````python
class Stack:
    def __init__(self):
        self.__items = []

    def push(self, item):
        self.__items.append(item)
    
    def pop(self):
        return self.__items.pop()

    def is_empty(self):
        return len(self.__items)==0

    def topValue(self):
        if not self.is_empty():
            return self.__items[-1]
        raise Exception("Stack is empty")
    
    def size(self):
        return len(self.__items)

    def items(self):
        return self.__items.copy()

values = [i for i in range(10)]
pila = Stack()
for v in values:
    pila.push(v)

print("Elementos de la pila:", pila.items())
topValue = pila.pop()
print("Elementos de la pila después de hacer pop:", pila.items())
print("Elemento eliminado con pop:", topValue)
````

## 4. Colas

````python
from queue import Queue

cola = Queue()
valuesToEnqueue = [i for i in range(10**6)]
for v in valuesToEnqueue:
    cola.put(v)

for i in range(cola.qsize()):
    cola.get() #dequeue encolar
cola.qsize()
````

## 5. Ejemplo Práctico

````python
import csv

def readFile(fileName='amazon_sales.csv'):
    with open(fileName, mode='r') as file:
        reader = csv.reader(file)
        columnsName = next(reader) # Leer la primera fila como encabezado
        data = [row for row in reader]  # Leer el resto de las filas como datos
        return columnsName, data
    
columnsName, data = readFile()
# print("columnsName:\n", columnsName)
print("First 5 rows of data:")
for i, row in enumerate(data[:5]):
    print(i, row)
````

````python
def getOnlyBestSellers(data, columnsName):
    bestSellers = []
    indexBestSeller = columnsName.index("is_best_seller")
    indexOriginalPrice = columnsName.index("original_price")
    for row in data:
        if row[indexBestSeller].lower() == "best seller" and\
            row [indexOriginalPrice] != "" : 
        # if row[indexBestSeller].lower() == "best seller":
            bestSellers.append(row)
    return bestSellers

def getAvgRating(data, columnsName):
    indexProductRating = columnsName.index("product_rating")
    avg = 0
    for row in data:
       avg += float(row[indexProductRating])
    return avg/len(data) 

def getAvgOriginalPrice(data, columnsName):
    indexOriginalPrice = columnsName.index("original_price")
    avg = 0
    for row in data:
        avg += float(row[indexOriginalPrice])
    return avg/len(data)



bestSellers = getOnlyBestSellers(data, columnsName)
print("Cantidad de Best Sellers:", len(bestSellers))
print(getAvgRating(bestSellers, columnsName))
print(getAvgOriginalPrice(bestSellers, columnsName))
````
