# Diccionarios Desordenados: Tablas de Hashing

Las tablas de hashing en python están implementadas en el clase `dict`:

* En particular implementa las tablas de hashing usando hashing cerrado,
* utilizando una política de resolución de colisiones basadas en una segunda función de hashing $h_2$, y
* utiliza un factor de carga $\alpha = \frac{2}{3}$

Como las utilizamos:
* Podemos definir una tabla de hashing vacía con las llaves: `ht = {}` o tambien con: `ht = dict()`
* Para agregar un par $(k, I)$ se puede hacer: `ht[k]=I` o con `setdeafault(k, I)`.
* Para eliminar un par $(k, I)$ se puede con: `del ht[k]`.

### Como crear pares (k, I)

```python
ht1 = {}
ht2 = dict()
ht1["hola"] = 1
ht1["mundo"] = [2, 4, 1, 5, 5]
# Si la clave no existe, se crea con el valor dado
ht1.setdefault("hola", 0)
ht1.setdefault("python", 0)
# Podemos agregar claves de cualquier tipo, 
# incluso mezclando tipos
ht1[1.1] = "x"

print(ht1)
print(ht2)
print("valor de la clave mundo:", ht1["mundo"])
```

### Como eliminar pares (k, I)

```python
ht = {"chile": "santiago", "peru": "lima", "argentina": "buenos aires"}
# Podemos eliminar claves con delete
del ht["chile"]
print("después de eliminar chile con del:", ht)

# Tambien podemos eliminar claves con pop, 
# pero esta función devuelve el valor eliminado
valor_eliminado = ht.pop("peru")
print("valor eliminado:", valor_eliminado)
print("después de eliminar peru con pop:", ht)

# Tambien podemos eliminar claves con popitem, 
# pero esta función elimina un par clave-valor al azar
clave_eliminada, valor_eliminado = ht.popitem()
print("clave, valor eliminados:", clave_eliminada, valor_eliminado)
```

## Concepto de hash

Para poder usar este tipo de diccionarios es importante entender que las claves tienen que se "hasheables". Por lo tanto, deben tener una función de hashing implementada para el  tipo de datos de la clave. Por ejemplo, en python existen funciones de hashing implementadas para strings, enteros, flotantes y tuplas(solo si son hasheables sus elementos). Para la información $I$, no hay problema, el tipo puede ser cualquiera.

Como podemos saber si algo es hasheable o no?, existe la función `hash()` la cual nos retornará el valor de

```python
hash(1)
hash(-1)
hash("hola mundo")
hash(( 1,2,3 ))
hash([1,2,3])
```

Los diccionarios usan la función `hash`, por lo tanto todo aquello que no sea hasheable para python no se puede utilizar como clave.

```python
ht = dict()
ht["mundo"] = 2
ht[1] = "info"
ht[1.1] = ["info2", ["info3"]]
```

#### Otras operaciones que Podemos hacer

```python
d = {
    "hola": 30,
    "mundo": 31,
    "python": 100
}
# Podemos obtener lo diferentes datos de la tabla
print(d.keys())
print(d.values())
print(d.items())

# Podemos preguntar si una llave está en la tabla
print("hola" in d) # O(1)
print("c++" in d) 

# Notar la diferencia entre preguntar si un valor está en una lista,
#  y preguntar si una llave está en una tabla hash
L =  ["mundo", "python", "hola"]
print("hola" in L) # O(n)
```

### Sets

Una cosa interesante de python es que la implementación del tipo `set` es utilizando tablas de hashing, por lo que hacer una consulta del tipo: `x in set` cuesta $O(1)$ vs las listas que cuesta $O(n)$. Es por esto tambien que no garantizan el orden de los elementos.

```python
S = set((19000, 2, 4, 5, 6, 8, 1)) # tablas de hashing (k, k)
print("2 in S:", 2 in S)

# Note como no preserva el orden original de los elementos
for e in S:
    print(e)
```

## Cuidado al usar tablas de hashing con objetos

```python
# clase X
class X:
    def __init__(self, value):
        self.value = value

objectX = X(1)
objectY = X(2)
objectZ = X(2)

print(objectX.value)
print(objectY.value)
print(objectZ.value)

ht = {}
ht[objectX] = [1]
ht[objectY] = [1,2]
ht[objectZ] = [1,2,3]

print(ht[objectY])
print(ht[objectZ])
```

Note como se crearon 3 datos en vez de sobre escribir el valor `objectY`

```python
ht[objectZ]
```

```python
import csv

def readFile(fileName=''):
    with open(fileName, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        columnsName = next(reader) # Leer la primera fila como encabezado
        data = [row for row in reader]  # Leer el resto de las filas como datos
        return columnsName, data

columnsName, data = readFile('FakeNews.csv')
print(columnsName)
print(data[0])
```

```python

#stop words
english_stopwords = [
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", 
    "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 
    'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it', 
    "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 
    'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 
    'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 
    'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 
    'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 
    'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 
    'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 
    'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 
    'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than', 
    'too', 'very', 's', 't', 'can', 'will', 'just', 'don', "don't", 'should', 
    "should've", 'now', 'd', 'll', 'm', 'o', 're', 've', 'y', 'ain', 'aren', 
    "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', 
    "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', 
    "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', 
    "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', 
    "wouldn't"
]
english_stopwords = set(english_stopwords)

def getOnlyFakeNews(data):
    fakeNews = []
    for row in data:
        if bool(row[-1]) and row[1] != '':
            fakeNews.append(row)
    return fakeNews

def splitText(text):
    return text.strip().split()

def getFrequences(data, columnsName):
    indexText = columnsName.index('text')
    wordsFrequence = {}
    for row in data:
        text = row[indexText]
        words = splitText(text)
        for word in words:
            auxWord = word.lower()
            if auxWord in english_stopwords:
                continue
            if auxWord in wordsFrequence:
                wordsFrequence[auxWord] += 1
            else:
                wordsFrequence[auxWord] = 1
    return wordsFrequence

def getMaxFrequence(wordsFrequence):
    max = 0
    topWord = ''
    for word, frequence in wordsFrequence.items():
        if frequence > max:
            topWord = word
            max = frequence
    return topWord, max


fakeNews = getOnlyFakeNews(data)
freq = getFrequences(fakeNews, columnsName)
listWordsFreq = list(freq.items())
listWordsFreq.sort(key=lambda x: x[1], reverse=True)
listWordsFreq[0:5]
# word, topFrequence = getMaxFrequence(freq)
# word, topFrequence
```
