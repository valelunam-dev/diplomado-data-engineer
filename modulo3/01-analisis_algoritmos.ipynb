## Ejemplo 1

```python
def promedio(lista):
    suma = 0.0
    n = len(lista) # Tamaño de la lista
    if n == 0: # retorna 0.0 si la lista es vacía
        return suma 
    for numero in lista: #itera n veces
        suma += numero
    return suma/n #retorna el promedio de los datos
```

### Cuál es el tiempo de ejecución del algoritmo promedio?
**Recordemos las operaciones básicas que contaremos**

|Operaciones|	Sintaxis en Python|
|-----------|---------------------|
|Asignación|	+=, *=, =|
|Aritméticas| 	+, *, -, +, /, //|
|Lógicas|	and, or, not|
|Relacionales|	<=, >=, <, >, !=, ==|
|Saltos|	return, break|

Note que la función promedio itera sobre los $n$ datos de la lista (es decir, por cada unos de los elementos de la lista), y suma los valores de cada uno de los elementos, para finalmente retornar el promedio. Si contamos cada una de las operaciones básicas, el tiempo de ejecución está dado por la función: $T(n) = 2\cdot n + 7$. Por lo tanto, el tiempo usando la notación aprendida es $O(n)$.

### Ejemplo 2

```python
def maximo(lista):
    n = len(lista) # Tamaño de la lista
    if n == 0:
        return # Si la lista es vacía retornamos
    maximo = lista[0] # definimos el máximo como el primer elemento
    for i in range(1, n): # iteramos desde el segundo elemento
        if lista[i] >  maximo:  # si encontramos un elemento mayor
            maximo = lista[i]   # lo actualizamos
    return maximo
```

### Cuál sería el tiempo de ejecución?

1. El algoritmo primero define el maximo como el primer elemento.
2. Luego itera sobre los $n-1$ datos restantes, buscando si existe un elemento más grande, y si lo hay actualiza la variable `maximo`.

Estamos interesados en entradas muy grandes (puesto que vamos a trabajar con grandes volumenes de datos), por lo tanto, podemos expresar el tiempo de ejecución como la siguiente función: $$T(n) = c\cdot n$$

Note que, cuando la entrada sea **muy grande** las constantes de la función serán despreciables. Por lo tanto el tiempo de ejecución usando la notación es $O(n)$.

### Ejemplo 3

```python
def find_duplicates(data): # Recibe una lista de datos
  n = len(data) # cantidad de datos
  for i in range(n): # iteramos sobre cada dato
    for j in range(n): # volvemos a iterar sobre cada dato
      if j != i and data[i] == data[j]:
        return True, data[i]
  return False, None 
```

```python
# Ejemplo de uso
numbers = [1, 5, 23, 5, 43, 8, 10]
duplicateExists, duplicateElement = find_duplicates(numbers)
if duplicateExists:
  print(f"Duplicate found: {duplicateElement}")
else:
  print("No duplicates found.")
```

### Cuál sería su tiempo de ejecución?
Note que la variable `i` toma los siguientes valores: `[0, 1, 2, ..., n-1]`, por lo que el `for` exterior realiza `n` iteraciones. Ahora, por cada valor que toma tenemos el siguiente comportamiento:
|valor de `i`| valores de `j`| número de iteraciones|
|------------|---------------|----------------------|
| `0`        | `[0, 1, 2, ..., n-1]`| `n`|
| `1`        | `[0, 1, 2, ..., n-1]`| `n`|
| `2`        | `[0, 1, 2, ..., n-1]`| `n`|
|`...`       | `...`| `...`|
| `n-1`        | `[0, 1, 2, ..., n-1]`| `n`|

Note que si contamos la catidad todal de iteraciones serían $n\cdot n$, lo cual nos da un tiempo de ejecución de $T(n) = c\cdot n^2$. Entonces, el tiempo de ejecución usando la notación sería $O(n)$

```python
from sys import maxsize # número muy grande

def sort(lista):
    n = len(lista) # largo de la lista
    listaOrdenada = []
    for i in range(0, n): # de i=0 a i=n-1
        min = maxsize
        minIndex = 0
        for j in range(0, n): # de j=0 a j=n-1
            if lista[j] != "x" and lista[j] < min:
                min = lista[j]
                minIndex = j
        lista[minIndex] = "x"
        listaOrdenada.append(min)
    
    return listaOrdenada
```

```python
# Ejemplo de utilización de la función sort
lista = [2, 5, 1, 4, 6, 10, 3]
print(sort(lista))
```
